```python
from trainingmonitor import TrainingMonitor
from sklearn.preprocessing import LabelBinarizer
from minivggnet import MiniVGGNet
from keras.optimizers import SGD
from keras.datasets import cifar10
import os
```

```python
# folder path for output
output = "output"
```

```python
#Laod CIFAR10 training and test data
((trainX, trainY), (testX, testY)) = cifar10.load_data()
trainX = trainX.astype("float") / 255.0
testX = testX.astype("float") / 255.0
```

```python
#covert labels to vectors
lb = LabelBinarizer()
trainY = lb.fit_transform(trainY)
testY = lb.transform(testY)
```

```python
#create label names
labelNames = ["airplane", "automobile", "bird", "cat", "deer",
	"dog", "frog", "horse", "ship", "truck"]
```

```python
# Create mdoel without learning rate (alpha)
opt = SGD(lr=0.01, momentum=0.9, nesterov=True)
model = MiniVGGNet.build(width=32, height=32, depth=3, classes=10)
model.compile(loss="categorical_crossentropy", optimizer=opt,
	metrics=["accuracy"])
```

```python
# create callbacks
figPath = os.path.sep.join([output, "{}.png".format(
	os.getpid())])
jsonPath = os.path.sep.join([output, "{}.json".format(
	os.getpid())])
callbacks = [TrainingMonitor(figPath, jsonPath=jsonPath)]
```

```python
#traing model
model.fit(trainX, trainY, validation_data=(testX, testY),
	batch_size=64, epochs=100, callbacks=callbacks, verbose=1)
```
